## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# A simple rule could classify pages using only one metric, such as whether clicks increased or decreased.
# However, page performance depends on multiple daily signals that change over time. A page may gain impressions while losing CTR, improve its ranking without increasing clicks, or experience temporary fluctuations.
# Machine learning can learn these patterns from historical observations instead of relying on fixed thresholds, making it better suited for predicting future page performance.
```
